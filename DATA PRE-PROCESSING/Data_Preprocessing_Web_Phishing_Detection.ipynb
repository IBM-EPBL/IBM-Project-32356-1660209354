**Handling ouliers**

```python
q1=df.domain_age.quantile(0.25)
q3=df.domain_age.quantile(0.75)
```

```python
IQR=q3-q1
```

```python
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5*IQR
```

```python
upper_limit
```

```python
df.median()
```

```python
df['domain_age']=np.where(df['domain_age']>upper_limit,30,df['domain_age'])
```
